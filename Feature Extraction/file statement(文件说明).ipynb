# file statement（文件说明）

The data set used in this paper comes from the 91-day desensitized DNS communication traffic from 2020-3-01 to
2020-05-31 on the large-scale real-world network provided by Qi’anxin Technology Group, involving 20512 domains and 431130 IP addresses,both have undergone encoding desensitization treatment. In addition, we also get the whois information of the domains and the geographic information of the ip addresses.Due to copyright issues, the original data set is not shown here, only a summary of the original data is provided.

本文使用的数据集来自于奇安信实验室（Qi An Xin Technology Group）提供的某真实大型网络 2020-3-01 到 2020-05-31 共 91 天的脱敏 DNS 通信流量，涉及20512个域和431130个IP地址，均经过编码脱敏处理。此外，我们还获取域的whois信息和ip地址的地理信息。同时我们拿到了域名的whois 信息及 ip 地址的地理信息包括国家、区域。由于版权问题，这里不展示原始数据集，仅提供原始数据的摘要。
