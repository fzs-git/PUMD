```python
#model_evaluate
path = r'Final_result\model_evaluate'
paths = os.listdir(path)
title_list = ['acc_model_evaluate', 'auc_model_evaluate', 'fscore_model_evaluate', 'gmean_model_evaluate', 'mcc_model_evaluate', 'precision_model_evaluate', 'recall_model_evaluate']
y_label_list = ['ACC','AUC','F1','gmean','MCC','Precision','Recall']
y_dict = dict(zip(title_list,y_label_list))
for i in paths:
    com_path = os.path.join(path,i)

    if os.path.isfile(com_path):
        df = pd.read_csv(com_path)
#         plt.rcParams["font.family"]="SimHei"
        display(df)
        
        col = ['label_frequency', 'PUMD_rf', 'PUMD_xgboost', 'PUMD_svm', 'PUMD_lr']
        rf = df[col[1]].to_list()
        xgboost = df[col[2]].to_list()
        svm = df[col[3]].to_list()
        lr = df[col[4]].to_list()
        x=np.arange(1,10)
        
        fig=plt.figure(num=1,figsize=(6,4))

        ax=fig.add_subplot(111)
        ax.plot(x,rf,"r-.d",label="RF")
        ax.plot(x,xgboost,"c-d",label="Xgboost")
        ax.plot(x,svm,"g--d",label="SVM")
        ax.plot(x,lr,"y:d",label="LR")
        
        label_freqency_list = np.arange(0,1,0.1)
        label_freqency_list = [round(i,2) for i in label_freqency_list]
        ax.set_xticklabels(label_freqency_list)
        print(i)
        title = i.split('.')[1].split('-')[-1]


        ax.set_xlabel("Label Frequency",fontsize=12)#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
        ax.set_ylabel(y_dict[title],fontsize=12)#添加y轴标签，设置字体大小为16，这里也可以设字体样式与颜色
        
        

        ax.legend(loc=3,labelspacing=0.5,handlelength=2,fontsize=10,shadow=False)
#         plt.savefig(r"D:\Project\Python\pu_test_3_18\5-22\fig\model_evaluate\\"+title+"model_evaluate_figure.svg", format="svg", dpi=600,bbox_inches='tight',pad_inches=0.0)
        plt.show()
        
```

## method_evaluate

```python
#method_evaluate
path = r'Final_result\method_evaluate'
paths = os.listdir(path)
for i in paths:
    com_path = os.path.join(path,i)
    if os.path.isfile(com_path):
        df = pd.read_csv(com_path)
        #df.style.highlight_max(axis=1,subset = pd.IndexSlice[0:9,'PU_ADOA_method':])
        print('csv_path:',com_path)
        display(df)
        df = df[['label_frequency','PUMD','PU_biased','PU_empirical','mix_supervised','unsupervised','pure_supervised']]
        begin = df.columns.to_list()[1]
        baseline = df.columns.to_list()[-3]
        print(begin,":",baseline)
        df['diff'] = np.array(df[begin].to_list())- np.array(df[ baseline].to_list())
        display(df.style.highlight_max(axis=1,subset = pd.IndexSlice[0:9,'PUMD':]))
        df.to_csv(com_path,index=False)
        
        
```

```python
#method_evaluate
path = r'Final_result\method_evaluate'
paths = os.listdir(path)
title_list = ['acc_method_evaluate', 'auc_method_evaluate', 'fscore_method_evaluate', 'gmean_method_evaluate', 'mcc_method_evaluate', 'precision_method_evaluate', 'recall_method_evaluate']
y_label_list = ['ACC','AUC','F1','gmean','MCC','Precision','Recall']
y_dict = dict(zip(title_list,y_label_list))
for i in paths:
    com_path = os.path.join(path,i)
    if os.path.isfile(com_path):
        df = pd.read_csv(com_path)
#         plt.rcParams["font.family"]="SimHei"
        
        col = ['label_frequency','PUMD','PU_biased','PU_empirical','mix_supervised','unsupervised','pure_supervised','diff']

        adoa = df[col[1]].to_list()
        biased = df[col[2]].to_list()
        emprical = df[col[3]].to_list()
        mix_supervised = df[col[4]].to_list()
        unsupervised = df[col[5]].to_list()
        pure_supervised = df[col[6]].to_list()
        x=np.arange(1,10)
        
        fig=plt.figure(num=1,figsize=(6,4))

        ax=fig.add_subplot(111)
        ax.plot(x,adoa,"r-.d",label="PUMD")
        ax.plot(x,biased,"c-d",label="PU_biased")
        ax.plot(x,emprical,"g--d",label="PU_emprical")
        ax.plot(x, mix_supervised,"y:d",label="mix_supervised")
        ax.plot(x,unsupervised,"k-d",label="unsupervised")
        ax.plot(x,pure_supervised,"m--d",label="pure_supervised")
        label_freqency_list = np.arange(0.1,1,0.1)
        label_freqency_list = [round(i,2) for i in label_freqency_list]
        ax.set_xticklabels(label_freqency_list)

        title = i.split('.')[1].split('-')[-1]
        ax.set_xlabel("Label Frequency",fontsize=12)#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
        ax.set_ylabel(y_dict[title],fontsize=12)#添加y轴标签，设置字体大小为16，这里也可以设字体样式与颜色
        
        
        ax.legend(loc=3,labelspacing=0.5,handlelength=2,fontsize=10,shadow=False)
#         plt.savefig(r"D:\Project\Python\pu_test_3_18\5-22\fig\method_evaluate\\"+title+"method_evaluate_figure.svg", format="svg", dpi=600,bbox_inches='tight',pad_inches=0.0)
        plt.show()
        
```

## org_append_selection

### test1

```python
path = r'Final_result\org_append_selection'
paths = os.listdir(path)
for i in paths:
    i = os.path.join(path,i)
    if "dl_append_fl" in i:
        df_append = pd.read_csv(i)  
    if "dl_selection_fl" in i:
        df_selection = pd.read_csv(i)
    if "dl_org_fl" in i:
        df_org = pd.read_csv(i)
```

```python
#org_append_selection

import math
columns=[ 'f_score', 'gmean', 'recall', 'precision', 'acc', 'auc', 'mcc']
y_label = ['F1','gmean','Recall','Precision','ACC','AUC','MCC']
y_dict = dict(zip(columns,y_label))
for col in columns:
    append_ = df_append[col].to_list()
    fs_ = df_selection[col].to_list()
    org_ = df_org[col].to_list()
    
#     plt.rcParams["font.family"]="SimHei"
    fig=plt.figure(num=1,figsize=(6,4))

    ax=fig.add_subplot(111)
    max_v = np.max(np.c_[ append_,fs_,org_])
    min_v = np.min(np.c_[ append_,fs_,org_])
    max_v = math.ceil(max_v*10)/10
    min_v = int(min_v*10)/10
    ax.set_ylim( min_v, max_v )
#   ax.set_yticks(np.arange(iint(min_v*10)/10,math.ceil(max_v*10)/10 ,(int(math.ceil(max_v*10)/10)-int(int(min_v*10)/10)/5)))
    ax.bar(x=np.arange(1,18,2),height=org_,color="firebrick",bottom=0,edgecolor="gold",label="Original",linewidth=2,width=0.8,alpha=1)
    ax.bar(x=np.arange(1+0.8,18+0.8,2),height=append_,color="b",bottom=0,edgecolor="gold",label="Feature_Append",linewidth=2,width=0.8,alpha=1)
    ax.bar(x=np.arange(1+0.8,18+0.8,2),height=list(np.array(fs_)-np.array(append_)),color="y",bottom=append_,edgecolor="gold",label="Feature_Selection",linewidth=2,width=0.8,alpha=1)
    ax.set_xticks(np.arange(1+0.4,18+0.4,2))
    label_freqency_list = np.arange(0.1,1,0.1)
    label_freqency_list = [round(i,2) for i in label_freqency_list]
    ax.set_xticklabels(label_freqency_list)
#     ax.xticks(ind, np.arange(0.1,1,0.1))
    ax.set_xlabel("Label Frequency",fontsize=12)#添加x轴坐标标签，后面看来没必要会删除它，这里只是为了演示一下。
    ax.set_ylabel(y_dict[col],fontsize=12)#添加y轴标签，设置字体大小为16，这里也可以设字体样式与颜色


    ax.legend(fontsize=10,shadow=False)
    plt.show()
```
